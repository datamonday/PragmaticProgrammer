## 求解斐波那契数列

```python
# 非常耗时，不推荐使用。
def fib(n):
    if n <= 0:
        return 0
    
    elif n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(10)
```

```python
# 改进1
def fib(first, second, n):
    if n <= 0:
        return 0
    
    if n <= 2:
        return 1
    
    elif n == 3:
        return first + second
    
    else:
        return fib(second, first+second, n - 1)
```

```python
fib(1, 1, 10)
```

```python
# 改进2
def fib(n):
    a, b = 1, 1
    for i in range(n-1):
        a, b = b, a + b
        
    return a
```

```python
fib(10)
```

```python
# 输出斐波那契数列
def fib(n):
    if n <= 0:
        return None
    elif n == 1:
        return [1]
    elif n == 2:
        return [1, 1]
    else:
        res = [1, 1]
        for i in range(2, n):
            res.append(res[-1]+res[-2])
            
    return res
```

```python
fib(10)
```

## 二分查找

```python
# 递归实现
def binary_search(od_list, key, left, right):
    if left > right:
        return None
    
    mid = (left + right) // 2
    if key == od_list[mid]:
        return mid
    elif key > od_list[mid]:
        return binary_search(od_list, key, mid+1, right)
    else:
        return binary_search(od_list, key, left, mid-1)
```

```python
test_list = [1, 2, 3, 4, 5, 8, 10]

binary_search(test_list, 8, 0, 6)
```

```python
# 非递归实现
def binary_search(od_list, key, left, right):
    while left < right:
        mid = (left + right) // 2
        if key == od_list[mid]:
            return mid
        elif key > od_list[mid]:
            left = mid + 1
        else:
            right = mid - 1
    return None
```

```python
test_list = [1, 2, 3, 4, 5, 8, 10]

binary_search(test_list, 8, 0, 6)
```
